```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('fraudTrain.csv', index_col=0)
df_test = pd.read_csv('fraudTest.csv', index_col=0)
```

```python
df_train.columns
```

```python
df_train.head()
```

```python
# Convert the trans_date_trans_time column to datetime
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])

# Create new columns for date and time
df_train['trans_date'] = df_train['trans_date_trans_time'].dt.date
df_train['trans_time'] = df_train['trans_date_trans_time'].dt.time

df_test['trans_date'] = df_test['trans_date_trans_time'].dt.date
df_test['trans_time'] = df_test['trans_date_trans_time'].dt.time

# Drop the trans_date_trans_time column
df_train.drop(['trans_date_trans_time'], axis=1, inplace=True)    
df_test.drop(['trans_date_trans_time'], axis=1, inplace=True)

# Display the first few rows to verify
df_train.head()
df_test.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=df_train, x='gender', hue='is_fraud', palette='coolwarm')

# Add labels and title
plt.title('Fraud vs Non-Fraud by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Is Fraud', loc='upper right')
plt.show()
```
